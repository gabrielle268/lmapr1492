```python
from pymatgen import Structure
from pymatgen.analysis.diffraction.xrd import XRDCalculator
from pymatgen.ext.matproj import MPRester
import matplotlib.pyplot as plt
with MPRester("2a0sQXpqwWhCvNGk") as m:
    structure = m.get_structure_by_material_id("mp-2784")
    structure.to(filename = "Na2Te.cif")
```

Pour déterminer les trois premiers pics du diffractogramme du Na2Te en utilisant un rayonnement X incident Cu Kα de longueur d'onde λ = 1,54060 Å, nous pouvons utiliser la formule de Bragg, qui est donnée par :

nλ = 2d sin(θ)

où n est un entier (le nombre d'ordre), λ est la longueur d'onde du rayonnement X, d est la distance entre les plans réticulaires, et θ est l'angle de diffraction. On obtient comme cela l'angle θ auxquels surviennent les pics.

Voici les étapes pour déterminer les trois premiers pics :

1.Trouver les indices hkl pour les plans réticulaires correspondant à chaque pic.
2.Calculer la distance interplanaire d pour chaque plan réticulaire.
3.Utiliser la formule de Bragg pour calculer l'angle de diffraction θ correspondant à chaque pic.
4.Utiliser les valeurs d'angle de diffraction pour identifier les trois premiers pics.

La formule générale pour la distance interplanaire d est donnée par :

d = a / √(h² + k² + l²)

où h, k, et l sont les indices de Miller qui caractérisent le plan cristallin.

```python
xrd = XRDCalculator('CuKa')
xrd_pattern = xrd.get_pattern(structure, scaled=True)
# premier pic du diffractogramme
print("1er pic du diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å")
print("Angle de diffraction (double) 2θ = ", xrd_pattern.x[0])
print("Intensité = ", xrd_pattern.y[0])
print("Indices hkl et multiplicité = ", xrd_pattern.hkls[0])
print("Distance associée :", xrd_pattern.d_hkls[0])
print("\n")
# deuxième pic du diffractogramme
print("2eme pic du diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å")
print("Angle de diffraction (double) 2θ = ", xrd_pattern.x[1])
print("Intensité = ", xrd_pattern.y[1])
print("Indices hkl et multiplicité = ", xrd_pattern.hkls[1])
print("Distance associée :", xrd_pattern.d_hkls[1])
print("\n")
# troisième pic du diffractogramme
print("3eme pic du diffractogramme pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å")
print("Angle de diffraction (double) 2θ = ", xrd_pattern.x[2])
print("Intensité = ", xrd_pattern.y[2])
print("Indices hkl et multiplicité = ", xrd_pattern.hkls[2])
print("Distance associée :", xrd_pattern.d_hkls[2])
print("\n")

xrd.show_plot(structure,annotate_peaks=True,two_theta_range=[0,40])
```
